```python
from sklearn.preprocessing import Binarizer
```

```python
binarizer = Binarizer(threshold =140)
```

```python
import pandas as pd
data = pd.DataFrame({"bp":[120,130,118,145,160]})
```

```python
data["bp_binary"] = binarizer.transform(data[['bp']])
```

```python
print(data)
```

```python
data['bp_label'] = data['bp'].apply(lambda x:"High" if x>=140 else "Normal")
```
