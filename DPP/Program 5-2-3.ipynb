```python
import pandas as pd

data = {
    "Transaction_Date": [
        "2023-07-14 09:32",
        "2023-08-02 18:45",
        "2023-09-21 11:10",
        "2023-10-05 22:05",
        "2023-11-30 07:50"
    ],
    "Customer_Review": [
        "Amazing service! Fast delivery üòä",
        "Delivery was late, but support helped.",
        "Quality is okay‚Ä¶ expected better üòê",
        "Worst experience ever!!! üò°üò°",
        "Will definitely recommend to others."
    ],
    "Membership_Type": [
        "Gold",
        "Silver",
        "Gold",
        "Silver",
        "Gold"
    ],
    "Notes_Raw_Data": [
        "###REF!@#$$",
        "$$$ERROR_404!!!",
        "@@NULL##DATA@@",
        "**##FAIL##**",
        "%%OK%%^&*()"
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.dtypes
```

```python
# Convert Transaction_Date to datetime
df["Transaction_Date"] = pd.to_datetime(df["Transaction_Date"])
df.dtypes
```

```python
#Extracting Date-Based Features
df['year'] = df['Transaction_Date'].dt.year
df['month'] = df['Transaction_Date'].dt.month
df['day'] = df['Transaction_Date'].dt.day
df['weekday'] = df['Transaction_Date'].dt.day_name()
df['hour'] = df['Transaction_Date'].dt.hour
```

```python
#display the newly added columns
print(df[df.columns[-5:]])
```

```python
#Feature Extraction from Customer Reviews (Text)
# Text length features
df["Review_Length"] = df["Customer_Review"].str.len()
df["Word_Count"] = df["Customer_Review"].str.split().str.len()

print(df["Review_Length"])
```

```python
# Sentiment Signal (Simple Emoji-Based Rule)
def sentiment_flag(text):
    if "üò°" in text:
        return -1   # Negative
    elif "üòä" in text:
        return 1    # Positive
    else:
        return 0    # Neutral

df["Sentiment_Flag"] = df["Customer_Review"].apply(sentiment_flag)
df["Sentiment_Flag"]
```

```python
#Feature Extraction from Membership Type (Categorical)
# Encode Membership Type
df["Is_Gold"] = (df["Membership_Type"] == "Gold").astype(int)
df["Is_Gold"]
```

```python
#Extract Signal from Noisy Notes Data
# Count special characters in Notes_Raw_Data
df["Special_Char_Count"] = df["Notes_Raw_Data"].str.count(r"[^a-zA-Z0-9]")
df["Special_Char_Count"]
```

```python
#Final Feature-Enriched Dataset
print(df)
```
