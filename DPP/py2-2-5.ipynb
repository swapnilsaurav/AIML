```python
import pandas as pd

data = pd.DataFrame({
    "Age": [55,47,48,45,32,44,57,52,50,37,43,45,50,43,31,57,58,43,53,40],
    "Salary":[101208, 110005,93909,122410,79584,83921,111923,54570,37349,55218,44236,65294,34144,40132,44093,54135,49116,113149,48081,127985]
})
print(data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

data['Age_scaled_minmax'] = scaler.fit_transform(data[['Age']])
data['Salary_scaled_minmax'] = scaler.fit_transform(data[['Salary']])
print(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data['Age_scaled_std'] = scaler.fit_transform(data[['Age']])
data['Salary_scaled_std'] = scaler.fit_transform(data[['Salary']])
print(data)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

data['Age_scaled_robust'] = scaler.fit_transform(data[['Age']])
data['Salary_scaled_robust'] = scaler.fit_transform(data[['Salary']])
print(data)
```
