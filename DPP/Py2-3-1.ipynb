```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({"income": [5, 8, 10, 12, 15, 20, 300, 500, 800]})

sns.histplot(df["income"], kde=True)
plt.show()

print("Skewness:", df["income"].skew())
```

```python
df['income_log'] = np.log1p(df['income'])

sns.histplot(df["income_log"], kde=True)
plt.show()

print("Skewness:", df["income_log"].skew())
```

```python
from scipy.stats import boxcox
df['income_boxcox'], lam = boxcox(df['income']+1)

sns.histplot(df["income_boxcox"], kde=True)
plt.show()

print("Skewness:", df["income_boxcox"].skew())
```

```python
from sklearn.preprocessing import PowerTransformer
pt= PowerTransformer(method="yeo-johnson")
df['income_yeoj']= pt.fit_transform(df[['income']])

sns.histplot(df["income_yeoj"], kde=True)
plt.show()

print("Skewness:", df["income_yeoj"].skew())
```
