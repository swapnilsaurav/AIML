<h2>Setting up Dataset</h2>

```python
#Python Demo – Dataset & Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.ensemble import IsolationForest
```

```python
#create a synthetic dataset with clusters.
X, _ = make_blobs(
    n_samples=300,
    centers=1,
    cluster_std=0.6,
    random_state=42
)
```

```python
#let’s manually add a few anomalies.
outliers = np.array([[4, 4], [-4, -4], [5, -5]])
X = np.vstack([X, outliers])
```

<h2>Training Isolation Forest</h2>

```python
#initialize the Isolation Forest model
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)

iso_forest.fit(X)
```

<h2>Predicting and Interpreting Results</h2>

```python
labels = iso_forest.predict(X)
scores = iso_forest.decision_function(X)
```

```python
#Visualizing Anomalies
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm')
plt.title("Isolation Forest – Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
