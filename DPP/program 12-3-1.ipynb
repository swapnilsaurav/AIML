```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import NearestNeighbors
```

```python
X, _ = make_blobs(
    n_samples=300,
    centers=1,
    cluster_std=0.6,
    random_state=42
)
```

```python
outliers = np.array([[4, 4], [-4, -4], [5, -5]])
X = np.vstack([X, outliers])
```

<h2>Computing kNN Distances </h2>

```python
k = 5
nbrs = NearestNeighbors(n_neighbors=k)
nbrs.fit(X)

distances, indices = nbrs.kneighbors(X)
```

```python
knn_score = distances[:, -1]
knn_score
```

```python
#Visualizing kNN Outlier Scores

plt.scatter(X[:, 0], X[:, 1], c=knn_score, cmap='viridis')
plt.colorbar(label='kNN Distance Score')
plt.title("kNN-Based Outlier Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
