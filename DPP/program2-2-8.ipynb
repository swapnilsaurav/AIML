```python
from sklearn.datasets import load_diabetes
import pandas as pd

diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)

print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

#MinMaxScaler
df_minmax = pd.DataFrame(MinMaxScaler().fit_transform(df), columns=df.columns)
print(df_minmax.head())
```

```python
#StandardScaler
df_stdscaler = pd.DataFrame(StandardScaler().fit_transform(df), columns=df.columns)
print(df_stdscaler.head())
```

```python
#RobustScaler
df_robust = pd.DataFrame(RobustScaler().fit_transform(df), columns=df.columns)
print(df_robust.head())
```
