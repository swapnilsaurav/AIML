```python
#Exercise: Feature Extraction from HR Data

import pandas as pd
from datetime import datetime

# Sample HR dataset
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Date_of_Joining": [
        "2018-06-15",
        "2020-01-10",
        "2019-09-01",
        "2021-03-22",
        "2017-11-30"
    ],
    "Last_Promotion_Date": [
        "2022-06-01",
        "2021-12-15",
        "2020-08-20",
        "2022-01-05",
        "2019-07-10"
    ],
    "Department": [
        "Sales",
        "Engineering",
        "HR",
        "Engineering",
        "Sales"
    ],
    "Performance_Rating": [4.5, 3.8, 4.0, 3.2, 4.7],
    "Monthly_Salary": [75000, 90000, 60000, 85000, 80000],
    "Overtime_Hours_Last_Month": [12, 5, 0, 20, 8],
    "Employee_Feedback": [
        "Love the team and growth opportunities",
        "Workload is high but learning a lot",
        "Supportive environment",
        "Too much pressure and deadlines",
        "Great leadership and culture"
    ]
}

df = pd.DataFrame(data)
df
```

```python
# Convert date columns to datetime
df["Date_of_Joining"] = pd.to_datetime(df["Date_of_Joining"])
df["Last_Promotion_Date"] = pd.to_datetime(df["Last_Promotion_Date"])

print(df)
```

```python
#Step 2: Date & Time Feature Extraction (HR-Relevant)
today = pd.Timestamp(datetime.now())

df["Tenure_Years"] = (today - df["Date_of_Joining"]).dt.days / 365
df["Years_Since_Last_Promotion"] = (today - df["Last_Promotion_Date"]).dt.days / 365

df
```

```python
#ðŸ· Step 3: Categorical Feature Extraction
df_encoded = pd.get_dummies(df, columns=["Department"], drop_first=True)

#Step 4: Behavioral & Numeric Feature Engineering
df_encoded["High_Overtime"] = df_encoded["Overtime_Hours_Last_Month"] > 10

#Salary Band Feature
df_encoded["Salary_Band"] = pd.cut(
    df_encoded["Monthly_Salary"],
    bins=[0, 65000, 80000, 100000],
    labels=["Low", "Medium", "High"]
)
df_encoded
```

```python
# Step 5: Text Feature Extraction
df_encoded["Feedback_Length"] = df_encoded["Employee_Feedback"].apply(len)

#Keyword Presence
df_encoded["Stress_Keyword"] = df_encoded["Employee_Feedback"].str.contains(
    "pressure|workload|deadlines",
    case=False
)
```

```python
#Final Feature-Engineered Dataset
print(df_encoded)
```
