```python
"""
Week 3 â€” Lesson 2, Video 2: Distance Measures (Euclidean, Manhattan, Cosine)
"""

import numpy as np

print("\n=== 2 Customers ===")

# Example: customer profile features (numeric)
# (Income in INR, SpendScore in 0-100, VisitsPerMonth)
cust_A = np.array([600000, 70, 6], dtype=float)
cust_B = np.array([900000, 65, 5], dtype=float)
```

```python

def euclidean(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def manhattan(a, b):
    return np.sum(np.abs(a - b))
```

```python
def cosine_distance(a, b):
    # Cosine distance = 1 - cosine similarity
    denom = (np.linalg.norm(a) * np.linalg.norm(b))
    if denom == 0:
        return 1.0
    cos_sim = np.dot(a, b) / denom
    return 1 - cos_sim
```

```python

print("Customer A:", cust_A)
print("Customer B:", cust_B)
print("Euclidean distance :", euclidean(cust_A, cust_B))
print("Manhattan distance :", manhattan(cust_A, cust_B))
print("Cosine distance    :", cosine_distance(cust_A, cust_B))
```
