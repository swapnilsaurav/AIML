```python
# Step 1- Import and setup
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
```

```python
#Load the data
# Load data
df = pd.read_csv("D:/datasets/dpp/curse-of-dim-demo-data.csv")

X = df.drop(columns=["target"]).values
y = df["target"].values
```

```python
# Identifying relevant features
# Relevant features are X1â€“X7
relevant_idx = list(range(7))
```

```python
#create train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# run the model with all the features
# === Model with ALL 20 features ===
rf_all = RandomForestRegressor(
    n_estimators=20,
    max_depth=4,
    max_features=5,
    random_state=42,
    n_jobs=-1
)
rf_all.fit(X_train, y_train)
y_pred_all = rf_all.predict(X_test)

r2_all = r2_score(y_test, y_pred_all)
rmse_all = mean_squared_error(y_test, y_pred_all) **0.5
print("R Square Score for all features: ",r2_all)
print("RMSE for all features: ",rmse_all)
```

```python
# === Model with ONLY 7 relevant features ===
X_train_rel = X_train[:, relevant_idx]
X_test_rel = X_test[:, relevant_idx]

rf_rel = RandomForestRegressor(
    n_estimators=20,
    max_depth=4,
    max_features=5,
    random_state=42,
    n_jobs=-1
)
rf_rel.fit(X_train_rel, y_train)
y_pred_rel = rf_rel.predict(X_test_rel)

r2_rel = r2_score(y_test, y_pred_rel)
rmse_rel = mean_squared_error(y_test, y_pred_rel) ** 0.5

print("RELEVANT FEATURES -> R2:", r2_rel, "RMSE:", rmse_rel)
```
